```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = 0
```

```python
df = pd.read_csv("data/contrato_convenio.csv")
df.head()
```

```python
df.columns
```

Cada entidade está representada por uma coluna no dataframe

```python
entidades = df.drop(columns=["arquivo_rast", "ato", "dodf", "treated_text", "IOB"])
print("Existem", len(entidades.columns), "entidades")
print("As entidades são:", entidades.columns)
```

Computa número de palavras por ato

```python
def num_palavras(x):
  if x is not np.nan: return len(x.split())
  else: return 0
```

```python
n_p = df.treated_text.apply(num_palavras)
```

```python
print("Quantidade total de atos:", len(n_p))
print("Número médio de palavras:", n_p.mean())
print("Número total de palavras:", n_p.sum())
```

Número médio e total de palavras por entidade

```python
df_temp = entidades.agg([lambda x: (x.apply(num_palavras)).mean(), lambda x: x.apply(num_palavras).sum()]).T
df_temp.columns = ["Media", "Total"]
df_temp.sort_values("Total", ascending=False)
```
