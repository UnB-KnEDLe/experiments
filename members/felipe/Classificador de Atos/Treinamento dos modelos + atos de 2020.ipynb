### Retira caracteres não-alfanuméricos ("(", ",", ".", "\n", etc)

```python
df.ato = df.ato.str.replace('\W', ' ')
df
```

### Remoção de stopwords

```python
stopwords = nltk.corpus.stopwords.words('portuguese')
```

```python
df['ato'] = df['ato'].apply(lambda x: ' '.join([item for item in x.split() if item not in stopwords]))
df
```

### Stemming

```python
stemmer = nltk.stem.RSLPStemmer()
```

```python
df['ato'] = df['ato'].apply(lambda x: ' '.join([stemmer.stem(item) for item in x.split()]))
df
```

### Aplicação do tfidf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words=stopwords, max_df = 0.9, min_df=2)
```

```python
X = vectorizer.fit_transform(df['ato'])
X
```

# Classificação 

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

from sklearn.svm import SVC # SVM
from sklearn.ensemble import RandomForestClassifier # Random Forest
from sklearn.neural_network import MLPClassifier # MLP
from sklearn.linear_model import LogisticRegression # Logistic Regression
from sklearn.naive_bayes import GaussianNB # Naive Bayes (Gaussian)
from sklearn.naive_bayes import ComplementNB # Naive Bayes (Complement)
from sklearn.tree import DecisionTreeClassifier # Decision Tree
from sklearn.linear_model import SGDClassifier # Stochastic Gradient Descent
```

```python
y = np.array(df['tipo_ato'])
```

```python
from sklearn.model_selection import StratifiedKFold
kf = StratifiedKFold(n_splits = 5, random_state = 87, shuffle = True)
```

### Função de Random Search

```python
def RSearchCV(clf, X, y, kfold = 5, iteracoes = 25, threads = -1, verbose = 0):

    classificadores = {
                       "SVM": (SVC(), [{'kernel': ['rbf'], 'gamma': ['auto', 'scale', 2, 1, 0.3, 0.1, 1e-2, 1e-3, 1e-4],
                                 'C': [0.5, 1, 2, 5, 30, 100, 300, 1000]},
                                 {'kernel': ['linear'], 'C': [0.5, 1, 2, 5, 30, 100, 300, 1000]}]),
                       
                       "RandomForestClassifier": (RandomForestClassifier(), {'max_features': ['sqrt', 'log2', 0.03, 0.1, 0.2, 0.4, 0.7], 
                                                   'n_estimators': [10, 25, 50, 100, 200, 300, 500, 1000, 1500],
                                                    'n_jobs': [2], 'random_state': [87]}),
                       
                       "MLP": (MLPClassifier(), {'solver': ['sgd', 'adam'], 
                                           'hidden_layer_sizes': [5, 10, 15, 20, (10, 10), (15, 10), (15, 15), (10, 10, 10)],
                                            'max_iter': [800, 1100, 1400, 1700], 'random_state': [87]}),
                       
                       "LogisticRegression": (LogisticRegression(), {'penalty': ['l1', 'l2', 'elasticnet'], 'C': [0.4, 0.8, 1.2, 1.6], 
                                               'solver': ['newton-cg', 'liblinear', 'saga']}),
                       
                       "GaussianNB": (GaussianNB(), {'var_smoothing': [1e-5, 1e-6, 1e-7, 1e-8, 1e-9, 1e-10]}),

                       "ComplementNB": (ComplementNB(), {'alpha': [0, 0.4, 0.8, 1.2, 1.6], 'norm': [True, False]}),

                       "DecisionTreeClassifier": (DecisionTreeClassifier(), {'criterion': ['gini', 'entropy'], 'random_state': [87], 
                                                    'class_weight': [None, 'balanced'], 'min_samples_split': [2, 10, 50, 100], 
                                                    'min_samples_leaf': [1, 5, 25]}),
                       
                       "SGDClassifier": (SGDClassifier(), {'loss': ['modified_huber', 'log'], 
                                          'penalty': ['l2', 'l1', 'elasticnet'], 'alpha': [0.00005, 0.0001, 0.0002], 'random_state': [87]})
    
    }
    
    score = 'f1_macro'
    classificador = classificadores[clf][0]
    parametros = classificadores[clf][1]
    
    clf = RandomizedSearchCV(classificador, param_distributions = parametros, n_iter = iteracoes, cv = kfold, random_state = 87, scoring = score, verbose = verbose, n_jobs = threads)
    clf.fit(X, y)
        
    return clf.best_estimator_
```

### Teste com o KFold

```python
index = 1
clfs_rf = []
for train, test in kf.split(X, y):
    print("Teste #%d" %index)
    X_train = X[train]
    X_test = X[test]
    y_train = y[train]
    y_test = y[test]
    
    clf = RSearchCV("RandomForestClassifier", X_train, y_train)
    clf.fit(X_train, y_train)
    clfs_rf.append(clf)
    
    y_true, y_pred = y_test, clf.predict(X_test)
    print(classification_report(y_true, y_pred), end = '\n\n')
    index += 1 # started at 00h37
```

```python
index = 1
clfs_dt = []
for train, test in kf.split(X, y):
    print("Teste #%d" %index)
    X_train = X[train]
    X_test = X[test]
    y_train = y[train]
    y_test = y[test]
    
    clf = RSearchCV("DecisionTreeClassifier", X_train, y_train)
    clf.fit(X_train, y_train)
    clfs_dt.append(clf)
    
    y_true, y_pred = y_test, clf.predict(X_test)
    print(classification_report(y_true, y_pred), end = '\n\n')
    index += 1
```

```python
index = 1
clfs_sgd = []
for train, test in kf.split(X, y):
    print("Teste #%d" %index)
    X_train = X[train]
    X_test = X[test]
    y_train = y[train]
    y_test = y[test]
    
    clf = RSearchCV("SGDClassifier", X_train, y_train)
    clf.fit(X_train, y_train)
    clfs_sgd.append(clf)
    
    y_true, y_pred = y_test, clf.predict(X_test)
    print(classification_report(y_true, y_pred), end = '\n\n')
    index += 1
```
