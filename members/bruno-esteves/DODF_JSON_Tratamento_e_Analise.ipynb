```python
path_JSONS = 'JSONS'

JSON_dict = {}
JSON_dict['JSON'] = []

list_JSON = os.listdir(path_JSONS)

for JSON in list_JSON:
  JSON_dict['JSON'].append(JSON)
```

***Tratamento dos JSON***

```python
import json
import re
```

```python
def pre_process(text):
  text = re.sub(r'<[^\>]*>', '', text)
  return text
```

*Análise dos Títulos e Tipos de Atos*

```python
TITULO_dict = {}
TIPO_dict = {}

DOCUMENT_dict = {}
DOCUMENT_dict['TIPO'] = []
DOCUMENT_dict['TITULO'] = []
DOCUMENT_dict['ATO'] = []


for JSON in JSON_dict['JSON']:
  try:

    f = open(path_JSONS + '/' + JSON)
    data = json.load(f)

    for ORGAO in data['json']['INFO']['Seção III']:
      for DOC in data['json']['INFO']['Seção III'][ORGAO]['documentos']:

        DOCUMENT_dict['TIPO'].append(data['json']['INFO']['Seção III'][ORGAO]['documentos'][DOC]['tipo'])
        DOCUMENT_dict['TITULO'].append(data['json']['INFO']['Seção III'][ORGAO]['documentos'][DOC]['titulo'])
        DOCUMENT_dict['ATO'].append(pre_process( data['json']['INFO']['Seção III'][ORGAO]['documentos'][DOC]['texto'] ))


        if data['json']['INFO']['Seção III'][ORGAO]['documentos'][DOC]['titulo'] in TITULO_dict:
          TITULO_dict[data['json']['INFO']['Seção III'][ORGAO]['documentos'][DOC]['titulo']] += 1
        else:
          TITULO_dict[data['json']['INFO']['Seção III'][ORGAO]['documentos'][DOC]['titulo']] = 1
        
        if data['json']['INFO']['Seção III'][ORGAO]['documentos'][DOC]['tipo'] in TIPO_dict:
          TIPO_dict[data['json']['INFO']['Seção III'][ORGAO]['documentos'][DOC]['tipo']] += 1
        else:
          TIPO_dict[data['json']['INFO']['Seção III'][ORGAO]['documentos'][DOC]['tipo']] = 1

    print(f"{JSON} Feito!")

    f.close()

  except:
    print(f"Erro na abertura do JSON: {JSON}")
```

***Análise de Tipos:***


1.   Licitação

*   ***Pode indicar a Reabertura de uma licitação passada, porém na maioria indica resultados de licitações***

2.   Contrato

*   ***Licitação já assinada ou algum aditivo de contrato***

3.   Pregão

*   ***Pode contar a indicação de abertura de uma nova licitação, mas na maior parte contém os resultados de licitações***

4.   Aviso de Licitação

*   ***Pode indicar um aviso de licitação fracassada, mas na maioria dos casos indica uma abertura ou futura abertura de licitação***

5.   Aviso

*   ***Na maioria dos casos indica uma abertura ou reabertura de licitação, mas pode ser também de resultado ou suspensão***

6.   Extrato

*   ***Pode indicar aditivos, dentre outros mas tem presente a maior parte das entidades presentes em uma licitação***
