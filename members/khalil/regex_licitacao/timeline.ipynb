```python
import pandas as pd
import glob
import re
```

```python
df_aberturas = pd.read_csv('abertura_regex.csv')
```

```python
txt_files = glob.glob('data/text/**/**/*.txt')
texto = ''
for txt in txt_files:
    texto += open(txt, encoding='utf-8').read()
```

```python
for index, row in df_aberturas.iterrows():
    if not pd.isna(row['processo']):
        
        p = re.compile(row['processo'])
        df_dict = {
            'local': [],
            'text': []
        }
        for m in p.finditer(texto):
            df_dict['local'].append(str(m.start()) + '-' + str(m.end()))
            df_dict['text'].append(texto[m.start()-2000: m.end() + 1000])
            print(m.start(), m.group())
        df = pd.DataFrame(df_dict)
        if len(df) != 1:
            df.to_csv('timeline_extraction/' + (m.group() + ".csv").replace('/', '-'), mode="w+")
```
