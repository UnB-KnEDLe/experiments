```python
import pandas as pd

arquivo = 'aditamento_contratual.csv'
df = pd.read_csv(arquivo)
num_atos = df.shape[0]

print(f'Existem {num_atos} atos')
df.head()
```

```python
colunas_removidas = ['treated_text', 'arquivo_rast', 'ato', 'dodf', 'IOB']  # colunas que não são entidades
entidades = df.drop(columns=colunas_removidas)
num_entidades = len(entidades.columns)

print('Os atos de aditamento contratual possuem as seguintes entidades: ')

for i in range(num_entidades):
    print(f'{i+1} - ' + entidades.columns[i])
```

```python
num_palavras_ato = [] # numero de palavras pra cada ato

for ato in df.treated_text:
    num_palavras_ato.append(len(ato.split()))

print(f'Número médio de palavras por ato: {sum(num_palavras_ato)//num_atos}')
```

```python
from numpy import nan

num_palavras_ent = [0] * num_entidades

for entidade in range(num_entidades):
    nome_entidade = entidades.columns[entidade]
    for ato in range(num_atos):  
        ent_ato = entidades[nome_entidade][ato]
        if ent_ato is not nan:
            num_palavras_ent[entidade] += len(ent_ato.split()) # incrementa a quantidade de palavras de cada entidade

dados = []
for i in range(num_entidades):
    entidade = entidades.columns[i]
    total = num_palavras_ent[i]
    media = total//num_atos
    dados.append([entidade, media, total])

df_aux = pd.DataFrame(data=dados, columns=['ENTIDADE', 'MÉDIA', 'TOTAL'])
df_aux.sort_values("MÉDIA", ascending=False)
```
