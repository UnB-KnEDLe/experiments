```python
import pandas as pd
import os
```

```python
data = pd.read_parquet("/home/gabriel/Desktop/unb/knedle/contract_extractor/df.parquet.gzip")
```

```python
data_grouped = data.groupby(["file_name", "number", "day", "month", "year"], as_index=False).first()
```

```python
del data
```

```python
data_grouped = data_grouped.drop(columns=["text", "page"])
```

```python
data_grouped = data_grouped.sort_values(["file_name", "year", "month", "day", "number"])
```

```python
samples_by_year = []
```

```python
for year in data_grouped["year"].unique():
    sample_by_year = data_grouped[data_grouped["year"] == year].sample(n=6, random_state=10).copy()
    sample_by_year["ns_year"] = range(1, len(sample_by_year)+1)
    samples_by_year.append(sample_by_year)
```

```python
samples_by_year_df = pd.concat(samples_by_year, ignore_index=True)
```

```python
samples_by_year_df = samples_by_year_df.sort_values(["year", "ns_year", "month", "day", "number", "file_name"]).reset_index(drop=True)
```

```python
path_dodfs_geral = "/home/gabriel/Desktop/unb/knedle/contract_extractor/accuracy_extractor/dodfs"
```

```python
dodfs_geral_years = os.listdir(path_dodfs_geral)
dodfs_geral_years.sort()
```

```python
dict_files_dodf = {
    "file": [],
    "year": [],
    "month": []
}

for year_path in dodfs_geral_years:
    path_year_months = os.path.join(path_dodfs_geral, year_path)
    files_path_year_months = os.listdir(path_year_months)
    
    for path_month in files_path_year_months:
        path_year_month_esp = os.path.join(path_year_months, path_month)
        files_path_month = os.listdir(path_year_month_esp)
        
        dict_files_dodf["file"] += files_path_month
        dict_files_dodf["year"] += [year_path]*len(files_path_month)
        dict_files_dodf["month"] += [path_month]*len(files_path_month)
```

```python
files_dodf_df = pd.DataFrame(dict_files_dodf)
```

```python
files_dodf_df["file_f"] = files_dodf_df["file"].str.split(".").str[0]
files_dodf_df["file_type"] = files_dodf_df["file"].str.split(".").str[-1]
```

```python
files_dodf_df.loc[files_dodf_df["file_f"] == files_dodf_df["file_type"], "file_type"] = "folder"
```

```python
files_dodf_df.head()
```

```python
samples_by_year_df["file_name"] = samples_by_year_df["file_name"].str.split("/").str[0]
```

```python
samples_by_year_df["file_name"].isin(files_dodf_df["file_f"]).all()
```

```python
path_files_original = "/home/gabriel/Desktop/unb/knedle/contract_extractor/accuracy_extractor/dodfs"
path_files_samples = "/home/gabriel/Desktop/unb/knedle/contract_extractor/accuracy_extractor/dodfs_samples"
```

```python
files_dodf_df_samples = files_dodf_df[files_dodf_df["file_f"].isin(samples_by_year_df["file_name"])].copy().reset_index(drop=True)
```

```python
files_dodf_df_samples = files_dodf_df.rename(columns={"file_f": "file_name", "year": "year_m", "month": "month_m"}).merge(samples_by_year_df, on=["file_name"], how="right")
```

```python
for year in files_dodf_df_samples["year_m"].unique():
    files_dodf_df_year_temp = files_dodf_df_samples[files_dodf_df_samples["year_m"] == year].copy().reset_index(drop=True)
    
    path_samples_year = os.path.join(path_files_samples, year)
    os.system("mkdir {}".format(path_samples_year))
    
    for month in files_dodf_df_year_temp["month_m"].unique():
        files_dodf_df_year_month_temp = files_dodf_df_year_temp[files_dodf_df_year_temp["month_m"] == month].copy().reset_index(drop=True)
        
        for file in files_dodf_df_year_month_temp["file"]:
            path_file_original = os.path.join(path_files_original, year, month)
            path_file_sample = path_samples_year
            n_file = files_dodf_df_year_month_temp[files_dodf_df_year_month_temp["file"] == file]["ns_year"].iloc[0]
            
            if ".pdf" in file:
                os.system("cp '{}/{}' '{}/{}_{}'".format(path_file_original, file, path_file_sample, n_file, file))
            else:
                os.system("cp -r '{}/{}' '{}/{}_{}'".format(path_file_original, file, path_file_sample, n_file, file))
```
