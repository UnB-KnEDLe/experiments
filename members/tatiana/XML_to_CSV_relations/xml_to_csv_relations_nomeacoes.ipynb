```python
import xml.etree.ElementTree as ET
tree = ET.parse('DODF10_nomeacoes_completo.xml')
root = tree.getroot()
```

```python
import os
import csv

csvFileNomeacaoComissionado = open('dodf10_nomeacaoComissionado.csv', 'w', encoding='utf-8')
csvNomeacaoComissionado = csv.writer(csvFileNomeacaoComissionado)

csvFileNomeacaoEfetivo = open('dodf10_nomeacaoEfetivo.csv', 'w', encoding='utf-8')
csvNomeacaoEfetivo = csv.writer(csvFileNomeacaoEfetivo)

#colNames = ['tipo', 'texto', 'NomeServidor', 'CargoEfetivo', 'Matricula', 'MatriculaSiape', 'Simbolo', 'Cargo', 'Hierarquia_da_Lotacao', 'Orgao']
#csvWriter.writerow(colNames)

colNomeacaoComissionado = ['id', 'annotator', 'tipo', 'texto', 'NomeServidor', 'CargoEfetivo', 'Matricula', 'MatriculaSiape', 'Simbolo', 'Cargo', 'Hierarquia_da_Lotacao', 'Orgao']
csvNomeacaoComissionado.writerow(colNomeacaoComissionado)

colNomeacaoEfetivo = ['id', 'annotator', 'tipo', 'texto', 'ProcessoSEI', 'Edital Normativo', 'Data do Edital Normativo', 'Número do DODF do Edital Normativo', 'Data do DODF do Edital Normativo', 'Edital de Resultado Final', 'Data do Edital de Resultado Final', 'Número do DODF do Edital de Resultado Final', 'Data do DODF do Edital de Resultado Final', 'Cargo', 'Carreira', 'Orgao', 'Especialidade', 'Candidatos (nome, classificacao)', 'Portador de Necessidades Especiais', 'Candidatos PNE (nome, Classificacao)']
csvNomeacaoEfetivo.writerow(colNomeacaoEfetivo)

for documento in root.findall('document'):
    #dentro do documento
    for paragrafo in documento.findall('passage'):
        #dentro de cada parágrafo
        for relation in paragrafo.findall('relation'):
            #dentro de cada relacao
            relationId = relation.get('id')
                    
            for infon in relation.findall('infon'):
                if infon.get('key') == 'annotator':
                    annotator = infon.text
            
            
            for infon in relation.findall('infon'):
                if infon.get('key') == 'type':
                    tipo = infon.text
                    
                    if tipo == 'NomeacaoCargoComissionado':
                        #dentro de cada anotação
                        anno_data = []
                        node_data = []
                        texto = 'Nan'
                        nome = 'Nan'
                        cargoEfetivo = 'Nan'
                        matricula = 'Nan'
                        matriculaSiape = 'Nan'
                        simbolo = 'Nan'
                        cargo = 'Nan'
                        hierarquia = 'Nan'
                        orgao = 'Nan'
                        print("\nRELACAO\n")

                        for node in relation.findall('node'):
                            annoId = node.get('refid')
                            node_data.append(annoId)

                        for i in node_data:

                            for anno in paragrafo.findall('annotation'):

                                if anno.get('id') == i:
                                    #print('achei anno ' + i)
                                    for info in anno.findall('infon'):           
                                        if info.get('key') == 'type':
                                            tipoAnno = info.text
                                            #print(tipoAnno)
                                    if tipoAnno == 'NomeacaoComissionado':
                                        texto = anno.find('text').text
                                        print(texto + '\n')
                                    elif tipoAnno == 'NomeServidor':
                                        nome = anno.find('text').text

                                    elif tipoAnno == 'CargoEfetivo':
                                        cargoEfetivo = anno.find('text').text

                                    elif tipoAnno == 'Matricula':
                                        matricula = anno.find('text').text

                                    elif tipoAnno == 'MatriculaSiape':
                                        matriculaSiape = anno.find('text').text

                                    elif tipoAnno == 'Simbolo':
                                        simbolo = anno.find('text').text

                                    elif tipoAnno == 'Cargo':
                                        cargo = anno.find('text').text

                                    elif tipoAnno == 'Hierarquia_da_Lotacao':
                                        hierarquia = anno.find('text').text

                                    elif tipoAnno == 'Orgao':
                                        orgao = anno.find('text').text
                        anno_data.append(relationId)
                        anno_data.append(annotator)
                        anno_data.append(tipo)               
                        anno_data.append(texto)
                        anno_data.append(nome)
                        anno_data.append(cargoEfetivo)
                        anno_data.append(matricula)
                        anno_data.append(matriculaSiape)
                        anno_data.append(simbolo)
                        anno_data.append(cargo)
                        anno_data.append(hierarquia)
                        anno_data.append(orgao)
                    
                        csvNomeacaoComissionado.writerow(anno_data)
                        print(node_data)
                        print(anno_data)
                                        
                    if tipo == 'NomeacaoCargoEfetivo':
                        #dentro de cada anotação
                        anno_data = []
                        node_data = []
                        listaCandidatos = []
                        listaCandidatosPNE = []
                        texto = 'Nan'
                        editalNormativo = 'Nan'
                        dataEditalNormativo = 'Nan'
                        numeroDODFEditalNormativo = 'Nan'
                        dataDODFEditalNormativo = 'Nan'
                        editalResultadoFinal = 'Nan'
                        dataEditalResultadoFinal = 'Nan'
                        numeroDODFEditalResultadoFinal = 'Nan'
                        dataDODFEditalResultadoFinal = 'Nan'
                        cargo = 'Nan'
                        carreira = 'Nan'
                        orgao = 'Nan'
                        especialidade = 'Nan'
                        nomeCandidato = 'Nan'
                        classificacao = 'Nan'
                        portadorNecessidadesEspeciais = 'nao'
                        nomePNE = 'Nan'
                        classificacaoPNE = 'Nan'
                        processoSEI = 'Nan'
                        print("\nRELACAO")

                        for node in relation.findall('node'):
                            annoId = node.get('refid')
                            node_data.append(annoId)

                        for i in node_data:

                            for anno in paragrafo.findall('annotation'):

                                if anno.get('id') == i:
                                    #print('achei anno ' + i)
                                    for info in anno.findall('infon'):           
                                        if info.get('key') == 'type':
                                            tipoAnno = info.text
                                            #print(tipoAnno)
                                    if tipoAnno == 'NomeacaoEfetivo':
                                        texto = anno.find('text').text
                                        print(texto + '\n')
                                    elif tipoAnno == 'EditalNormativo':
                                        editalNormativo = anno.find('text').text
                                        
                                    elif tipoAnno == 'DataEditalNormativo':
                                        dataEditalNormativo = anno.find('text').text
                                        
                                    elif tipoAnno == 'NumeroDODF_EditalNormativo':
                                        numeroDODFEditalNormativo = anno.find('text').text
                                        
                                    elif tipoAnno == 'DataDODF_EditalNormativo':
                                        dataDODFEditalNormativo = anno.find('text').text

                                    elif tipoAnno == 'EditalResultadoFinal':
                                        editalResultadoFinal = anno.find('text').text

                                    elif tipoAnno == 'DataEditalResultadoFinal':
                                        dataEditalResultadoFinal = anno.find('text').text    
                                    
                                    elif tipoAnno == 'NumeroDODF_ResultadoFinal':
                                        numeroDODFResultadoFinal = anno.find('text').text

                                    elif tipoAnno == 'DataDODF_ResultadoFinal':
                                        dataDODFEditalResultadoFinal = anno.find('text').text

                                    elif tipoAnno == 'Cargo':
                                        cargo = anno.find('text').text

                                    elif tipoAnno == 'ProcessoSei':
                                        processoSEI = anno.find('text').text  
                                        
                                    elif tipoAnno == 'Carreira':
                                        carreira = anno.find('text').text
                                        
                                    elif tipoAnno == 'Orgao':
                                        orgao = anoo.find('text').text
                                        
                                    elif tipoAnno == 'Especialidade':
                                        especialidade = anno.find('text').text
                                        
                                    elif tipoAnno == 'NomeServidor':
                                        nomeCandidato = anno.find('text').text
                                        if portadorNecessidadesEspeciais == 'nao':
                                            listaCandidatos.append(nomeCandidato)
                                        elif portadorNecessidadesEspeciais == 'sim':
                                            listaCandidatosPNE.append(nomeCandidato)
                                        
                                    elif tipoAnno == 'Classificacao':
                                        classificacao = anno.find('text').text
                                        if portadorNecessidadesEspeciais == 'nao': 
                                            listaCandidatos.append(classificacao)
                                        elif portadorNecessidadesEspeciais == 'sim':
                                            listaCandidatosPNE.append(classificacao)
                                            
                                    elif tipoAnno == 'PortadorNecessidadesEspeciais':
                                        portadorNecessidadesEspeciais = 'sim'                    
                                        
                        anno_data.append(relationId)
                        anno_data.append(annotator)
                        anno_data.append(tipo)               
                        anno_data.append(texto)
                        anno_data.append(processoSEI)
                        anno_data.append(editalNormativo)
                        anno_data.append(dataEditalNormativo)
                        anno_data.append(numeroDODFEditalNormativo)
                        anno_data.append(dataDODFEditalNormativo)
                        anno_data.append(editalResultadoFinal)
                        anno_data.append(dataEditalResultadoFinal)
                        anno_data.append(numeroDODFEditalResultadoFinal)
                        anno_data.append(dataDODFEditalResultadoFinal)
                        anno_data.append(cargo)
                        anno_data.append(carreira)
                        anno_data.append(orgao)
                        anno_data.append(especialidade)
                        anno_data.append(listaCandidatos)
                        #anno_data.append(classificacao)
                        anno_data.append(portadorNecessidadesEspeciais) 
                        anno_data.append(listaCandidatosPNE)
                        #anno_data.append(classificacaoPNE)
                    
                        csvNomeacaoEfetivo.writerow(anno_data)
                        print(node_data)
                        print(anno_data)
                        
                    
csvFileNomeacaoComissionado.close()
csvFileNomeacaoEfetivo.close()
```
