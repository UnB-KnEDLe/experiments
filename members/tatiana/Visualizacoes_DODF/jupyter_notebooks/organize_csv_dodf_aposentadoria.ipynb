# Organizando o conjunto de dados DODF_Aposentadoria.xlsx

O conjunto de dados original está disponível no drive do projeto KnEDLe como o nome 'DODF_Atos_de_Aposentadoria_validados_201804_201909.xlsx'.

Para facilitar o processamento dos dados e suas visualizações, eu adicionei a coluna 'int_label', 'conteudo' e 'label' ao dataset.

## 1. Importar biblioteca pandas e conjunto de dados

```python
import pandas as pd

data = pd.read_excel("DODF_Aposentadoria.xlsx")
data.head()

data
```

## 2. Criar coluna 'conteudo'

### 2.1. Gerar lista 'conteudo' a partir da concatenação dos dados de cada linha

Para criar a coluna 'conteudo', peguei os dados de cada coluna de uma linha e juntei em uma string. Depois, eu adicionei a string resultante à lista 'conteudo'

```python
conteudo = []
colunas = ["REF_ANOMES", "DATA_DODF", "NUM_DODF", "PAGINA_DODF", "TIPO_DODF", 'ATO', 'EMPRESA_ATO', 'COD_MATRICULA_ATO', 'COD_MATRICULA_SIGRH', 'CPF', 'NOME_ATO', 'CARGO', 'CLASSE', 'PADRAO', 'QUADRO', 'PROCESSO', 'FUND_LEGAL']
string = ""

i = 0
while i < len(data):
    for coluna in colunas:
        X = data[coluna][i]
        string = string + str(X) + ' '
    
    
    conteudo.append(string)
    string = ''
    i = i + 1
    
len(conteudo)
```

### 2.2. Criar coluna 'conteudo' no data frame

Depois criei a coluna 'conteudo' e adicionei os dados contidos na lista 'conteudo'

```python
data['conteudo'] = conteudo
data
```

## 3. Criar coluna 'label'

### 3.1 Definir o que será a 'label' e listar todas as labels presentes no conjunto de dados

Primeiro, decidi que a label seria a coluna 'EMPRESA_ATO' para seguir o padrão do conjunto de dados 'dodftrain.csv' onde as labels são os departamento do GDF responsáveis por cada ato. 

Depois de decidir a label, precisei descobrir quais eram para poder gerar o dicionário que geraria a lista 'int_label'. Para tal, eu passei as labels para a lista 'labels', evitando repetições. Depois, criei a lista 'labels_organizado', que é a lista 'labels' só que em ordem alfabética, para fins de verificar se ainda tinham repetições ou coisas do tipo. 

```python
label = data['EMPRESA_ATO']
labels =[]

for i in label:
    if i not in labels:
        labels.append(i)
        
labels_organizado = sorted(labels)

labels_organizado
```
