# Conversor de arquivos XML para CSV

O objetivo deste código é pegar as informações contidas nas anotações presentes em arquivos XML que foram gerados pelo TeamTat e passá-las para arquivos CSV de acordo com cada ato.

```python
def acharEntidades(tipo, dictAto, dictGeral, csv, dictAtoLista):
    #dentro de cada anotação
    for entidade in dictAto:
        dictAto[entidade] = 'nan'           

    for entidade in dictAtoLista:
        dictAtoLista[entidade] = []
            
    node_data = [] 
    for node in relation.findall('node'):
        annoId = node.get('refid')
        node_data.append(annoId)

    for i in node_data:
        for anno in paragrafo.findall('annotation'):
            if anno.get('id') == i:             
                for info in anno.findall('infon'):           
                    if info.get('key') == 'type':
                        tipoAnno = info.text
                                            
                for entidade in dictAto:
                    if tipoAnno == entidade:
                        dictAto[entidade] = anno.find('text').text           
                                
                for entidade in dictAtoLista:
                    if tipoAnno == entidade:
                        dictAtoLista[entidade].append(anno.find('text').text)
                            
    anno_data = [] 
    for item in dictGeral:
        escrever = dictGeral[item]
        anno_data.append(escrever)
                            
    for item in dictAto:
        escrever = dictAto[item]
        anno_data.append(escrever)
        
    for item in dictAtoLista:
        if dictAtoLista[item] == []:
            escrever = 'nan'
        else:
            escrever = dictAtoLista[item]
            anno_data.append(escrever)
                    
    csv.writerow(anno_data)

for root in roots:
    for documento in root.findall('document'):
        #dentro do documento
        for paragrafo in documento.findall('passage'):
            #dentro de cada parágrafo
            for relation in paragrafo.findall('relation'):
                #dentro de cada relacao

                dictGerais['id'] = relation.get('id')

                for infon in relation.findall('infon'):
                    if infon.get('key') == 'annotator':
                        dictGerais['anotador'] = infon.text  

                for infon in relation.findall('infon'):
                    if infon.get('key') == 'type':
                        tipo = infon.text
                        dictGerais['tipo'] = infon.text   

                        if tipo == 'Ato_Abono_Permanencia':
                            acharEntidades(tipo, dictAbonoPermanencia, dictGerais, csvAbonoPermanencia, [])

                        elif tipo == 'Ato_Aposentadoria':
                            acharEntidades(tipo, dictAposentadoria, dictGerais, csvAposentadoria, [])
                            
                        elif tipo == 'Ato_Cessao':
                            acharEntidades(tipo, dictCessao, dictGerais, csvCessao, dictCessaoLista)

                        elif tipo == 'Ato_Exoneracao_Cargo_Comissionado':
                            acharEntidades(tipo, dictExoneracaoComissionado, dictGerais, csvExoneracaoComissionado, [])

                        elif tipo == 'Ato_Exoneracao_Cargo_Efetivo':
                            acharEntidades(tipo, dictExoneracaoEfetivo, dictGerais, csvExoneracaoEfetivo,[])

                        elif tipo == 'Ato_Nomeacao_Cargo_Comissionado':
                            acharEntidades(tipo, dictNomeacaoComissionado, dictGerais, csvNomeacaoComissionado,[])

                        elif tipo == 'Ato_Nomeacao_Cargo_Efetivo':
                            acharEntidades(tipo, dictNomeacaoEfetivo, dictGerais, csvNomeacaoEfetivo, dictNomeacaoEfetivoLista)
                            
                        elif tipo == 'Ato_Retificacao_Cargo_Comissionado':
                            acharEntidades(tipo, dictRetificacaoComissionado, dictGerais, csvRetificacaoComissionado,[])

                        elif tipo == 'Ato_Retificacao_Cargo_Efetivo':
                            acharEntidades(tipo, dictRetificacaoEfetivo, dictGerais, csvRetificacaoEfetivo,[])
                            
                        elif tipo == 'Ato_Reversao':
                            acharEntidades(tipo, dictReversao, dictGerais, csvReversao,[])

                        elif tipo == 'Ato_Substituicao':
                            acharEntidades(tipo, dictSubstituicao, dictGerais, csvSubstituicao, dictSubstituicaoLista)

                        elif tipo == 'Ato_Tornar_Sem_Efeito_Apo':
                            acharEntidades(tipo, dictTornadoSemEfeitoApo, dictGerais, csvTornadoSemEfeitoApo, [])
                            
                        elif tipo == 'Ato_Tornado_Sem_Efeito_ExoNom':
                            acharEntidades(tipo, dictTornadoSemEfeitoExoNom, dictGerais, csvTornadoSemEfeitoExoNom, [])
                        
for file in csvFiles:
    file.close()
```
